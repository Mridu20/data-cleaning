**Encoding**

```python
import pandas as pd
```

3.Label encoding

```python
df = pd.DataFrame({'name':['cow','dog','green','green','red','blue']})
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["name"] = le.fit_transform(df["name"])
```

```python
df
```

4.Ordinal encoding

```python
dt = pd.DataFrame({'size':['s','m','l','xl','l','s', 'm', 's', 's', 'xl']})
```

```python
dt
```

```python
or_dt = [['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ore = OrdinalEncoder(categories= or_dt)
```

```python
ore.fit(dt)
```

```python
dt['en_size'] = ore.transform(dt)
```

```python
dt
```
