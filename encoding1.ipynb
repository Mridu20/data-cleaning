```python
import pandas as pd
```

```python
df = pd.read_csv("/content/CollegePlacement.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
en_df = df[["Internship_Experience", "Placement"]]
```

```python
pd.get_dummies(en_df)
```

```python
pd.get_dummies(en_df).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ar = ohe.fit_transform(en_df).toarray()
```

```python
pd.DataFrame(ar, columns = ["Internship_Experience_No", "Internship_Experience_Yes", "Placement_No", "Placement_Yes"])
```

```python
ohe2 = OneHotEncoder(drop = "first")
arr = ohe2.fit_transform(en_df).toarray()
```

```python
pd.DataFrame(arr, columns = [ "Internship_Experience_Yes", "Placement_Yes"])
```
